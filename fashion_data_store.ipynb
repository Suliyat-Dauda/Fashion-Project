```python
from ast import literal_eval

att_lists = fd1['p_attributes'].tolist()
id_lists = fd1['p_id'].tolist()
```

```python
def expand(chosen_column, join_id_column,desired_column_name):

    dictionary = []

    for i,k in zip(chosen_column,join_id_column):
        if i:
            i = literal_eval(i)
            dict1 = {desired_column_name: k }
            new_dict = {**dict1, **i}
            dictionary.append(new_dict)

    return pd.DataFrame(dictionary)
```

```python
def avg_rating_range(i: float) -> str:
    if 0.0 <= i < 1.0:
        return "1 star"
    elif 1.0 <= i < 2.0:
        return "2 stars"
    elif 2.0 <= i < 3.0:
        return "3 stars"
    elif 3.0 <= i < 4.0:
        return "4 stars"
    else:
        return "5 stars"
    
fd1_clean['avg_rating_range'] = fd1_clean_viz['avg_rating'].apply(avg_rating_range)

```
